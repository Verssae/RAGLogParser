```python
from pathlib import Path
import pandas as pd

stats = pd.read_csv("output/stats.csv")
stats
```

```python
projects = ["Android", "BGL", "Hadoop", "HealthApp", "Linux", "Mac", "Thunderbird" ]
shots = [0, 1, 2, 4]

        
print(f"{'project':12} | logs  | #tem  | {0:^13} | {1:^13} | {2:^13} | {4:^13} |") 
print(f"{'':12} |       |       |", end=" ") 
print(f"{'GA':^5} | {'PA':^5} | " * len(shots))
print("-" * 94)
for project in projects:
    print(f"{project:12} | ", end="") 
    df = pd.read_csv(f"output/{project}_0.csv")
    print(f"{len(df):5} | {len(df['template'].unique()):5}", end=" | ") 
    rows = stats[stats["project"] == project]
    for shot in shots:
        row = rows[rows["shot"] == shot].iloc[0]
        print(f"{row['GA']:^.3f} | {row['PA']:^.3f}", end=" | ")
    print()
```
